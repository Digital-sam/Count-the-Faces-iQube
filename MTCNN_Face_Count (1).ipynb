# Count The Faces

A computer vison(CV) model is a processing block that takes uploded inputs, like images or videos and predicts or returns pre-learned concepts or lebels. CV models can be trained to see almost anything humans can see. The general model CV can recognize concepts and give it a meaning.

#### There are various approches to solving a CV problem namely:
<ul>
    <li>Building a model from scratch</li>
    <li>The use of pretrained model and fine tune for the desired purpose</li>
</ul>

In this project, the approach to solving this problem is to use a Multi-task Cascaded Convolutional Neural Network(MTCNN)  which is a neural network that detects faces and facial landmarks on images. It was published by Zhang et al "Joint Face detection and alignment using multitask Cascaded Convolutional Network" in 2016. The major advantage of using the MTCNN is the ability to run well and with great accuracy on CPU. The MTCNN is one of the most popular and most accurate face detection tools today.

```python
# Loading required libraries
from PIL import Image
from numpy import asarray
from mtcnn.mtcnn import MTCNN 
import tensorflow as tf
import tensorflow.keras as Keras
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
#Create a directory

filename = 'C:/Users/LENOVO/Desktop/iQube/test_'
```

```python
#loading sample image
dir = filename + '/' + "tri.jpg"           # loading image
image = Image.open(dir)                    #loading image from given path
image = image.convert('RGB')               # converting image into 3D (RGB) image
pixels = asarray(image)

plt.imshow(pixels)
plt.axis('off')
pixels.shape
```

```python
#extraction using MTCNN
detector = MTCNN()     
results = detector.detect_faces(pixels)

total_faces = len(results)

for face in range(total_faces):
  print('box : {}'.format(results[face]['box']))
  print('confidence : {}'.format(results[face]['confidence']))
  print('keypoints : {}'.format(results[face]['keypoints']))
```

```python
# Use the MTCNN model to count the number of faces

output = []

# diretory of image
dir = filename + '/' + "truu.jpg"
image = Image.open(dir)            #loading image from given path
image = image.convert('RGB')       # converting image into 3D (RGB) image
pixels = asarray(image)            


# face extraction
detector = MTCNN()                 
results = detector.detect_faces(pixels)

# len(results) returns number of faces in an image
output.append(len(results))
```

```python
#show the image and the prediction
plt.imshow(pixels)
print('number of people in the picture is : {}'.format(output))
```
